# 1.Import Libraries

```python
import os
import cv2
import random
from tqdm import tqdm
```

# 2.Paths

```python
video_path = "/mnt/d/deepfake/deepfake-env/FF++"
output_path = "/mnt/d/deepfake/tf-gpu-env/complete_project/frames"

blur_threshold = 100
```

# 3. Blur Detection

```python
def is_blurry(frame,threshold):
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    variance=cv2.Laplacian(gray,cv2.CV_64F).var()

    return variance < threshold
```

# 4. Random Frame Extraction

```python
def extract_frames(video_file,label):
    cap=cv2.VideoCapture(video_file)
    video_name=os.path.basename(video_file).split(".")[0]
    frame_id=0
    while True:
        ret,frame=cap.read()
        if not ret:
            break
        if is_blurry(frame,blur_threshold):
            frame_id+=1
            continue
        save_name=f"{video_name}_{frame_id}.jpg"
        save_path=os.path.join(output_path,label,save_name)
        cv2.imwrite(save_path,frame)
        frame_id+=1
    cap.release()
```

# 5.Process Dataset

```python
for label in ["real","fake"]:
    folder=os.path.join(video_path,label)
    videos=os.listdir(folder)
    print(f"Processing {label} videos...")
    for video in tqdm(videos):
        video_file=os.path.join(folder,video)
        extract_frames(video_file,label)
```

# 6.Real and Fake Frames

```python
real_frames = len(os.listdir("/mnt/d/deepfake/tf-gpu-env/complete_project/frames/real"))
fake_frames = len(os.listdir("/mnt/d/deepfake/tf-gpu-env/complete_project/frames/fake"))

print("Real frames:", real_frames)
print("Fake frames:", fake_frames)
print("Total frames:", real_frames + fake_frames)
```
