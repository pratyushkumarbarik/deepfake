# 1.Import Libraries

```python
import os
import random
import shutil
from tqdm import tqdm
```

# 2.Paths

```python
frames_path="/mnt/d/deepfake/tf-gpu-env/complete_project/frames"
faces_path="/mnt/d/deepfake/tf-gpu-env/complete_project/faces"

frames_split="/mnt/d/deepfake/tf-gpu-env/complete_project/frames_train_test"
faces_split="/mnt/d/deepfake/tf-gpu-env/complete_project/faces_train_test"
```

# 3. Create Sub-folders

```python
def create_folders(base):

    for split in ["train","val"]:
        for label in ["real","fake"]:
            path=os.path.join(base,split,label)
            os.makedirs(path,exist_ok=True)

create_folders(frames_split)
create_folders(faces_split)
```

# 4.Dataset Split

```python
def split_dataset(source,target,ratio=0.8):

    images=os.listdir(source)
    random.shuffle(images)

    split=int(len(images)*ratio)

    train_images=images[:split]
    val_images=images[split:]

    label=os.path.basename(source)

    print(f"Copying train {label} images...")

    for img in tqdm(train_images):

        src=os.path.join(source,img)
        dst=os.path.join(target,"train",label,img)

        shutil.copy(src,dst)

    print(f"Copying val {label} images...")

    for img in tqdm(val_images):

        src=os.path.join(source,img)
        dst=os.path.join(target,"val",label,img)

        shutil.copy(src,dst)

    print(label,"train:",len(train_images),"val:",len(val_images))
```

# 5.Frame train_test

```python
for label in ["real","fake"]:

    source=os.path.join(frames_path,label)

    split_dataset(source,frames_split)
```

# 6. Faces train_test

```python
for label in ["real","fake"]:

    source=os.path.join(faces_path,label)

    split_dataset(source,faces_split)
```
