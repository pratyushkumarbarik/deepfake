```python
import torch
import torch.nn as nn
import timm
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt

from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.image import show_cam_on_image
from torchvision import transforms
```

```python
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
class DeepfakeModel(nn.Module):

    def __init__(self):

        super().__init__()

        self.cnn = timm.create_model(
            "efficientnet_b0",
            pretrained=False,
            num_classes=0
        )

        self.lstm = nn.LSTM(
            input_size=1280,
            hidden_size=256,
            num_layers=1,
            batch_first=True
        )

        self.fc = nn.Linear(256,1)

    def forward(self,x):

        B,T,C,H,W = x.shape

        x = x.view(B*T,C,H,W)

        features = self.cnn(x)

        features = features.view(B,T,-1)

        lstm_out,_ = self.lstm(features)

        out = lstm_out[:,-1,:]

        out = self.fc(out)

        return out
```

```python
model = DeepfakeModel().to(DEVICE)

model.load_state_dict(
    torch.load(
        "/mnt/d/deepfake/tf-gpu-env/complete_project/models/deepfake_model_temporal.pth",
        map_location=DEVICE
    )
)

model.eval()
```

```python
target_layer = model.cnn.blocks[-2]
```

```python
cam = GradCAM(
    model=model.cnn,
    target_layers=[target_layer]
)
```

```python
transform = transforms.Compose([
    transforms.Resize((160,160)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485,0.456,0.406],
        std=[0.229,0.224,0.225]
    )
])
```

```python
image_path = "/mnt/d/deepfake/tf-gpu-env/complete_project/faces_train_test/val/fake/01_20__outside_talking_pan_laughing__OTGHOG4Z_109_0.jpg"

img = Image.open(image_path).convert("RGB")

img = img.resize((160,160))

img_np = np.array(img) / 255.0
```

```python
input_tensor = transform(img).unsqueeze(0).to(DEVICE)
```

```python
grayscale_cam = cam(input_tensor=input_tensor)

heatmap = grayscale_cam[0]
```

```python
visualization = show_cam_on_image(
    img_np,
    heatmap,
    use_rgb=True,
    image_weight=0.75
)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(visualization)
plt.title("Grad-CAM Explanation")
plt.axis("off")
plt.show()
```

```python
heatmap_norm = cv2.normalize(
    heatmap,
    None,
    0,
    255,
    cv2.NORM_MINMAX
).astype(np.uint8)

_, thresh = cv2.threshold(
    heatmap_norm,
    180,
    255,
    cv2.THRESH_BINARY
)
```
